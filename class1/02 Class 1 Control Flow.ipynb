# Control Flow ควบคุมการทำงานของโปรแกรม

ในการเขียนคำสั่งต่างๆ เมื่อเราต้องการนำคำสั่งมาเรียบเรียงเป็นขั้นตอนที่เฉพาะเจาะจง หลายครั้งที่แค่นำแต่ละคำสั่งมาเรียงต่อกันเป็นขั้นตอนไม่เพียงพอ เช่น ในการสร้างเงื่อนไขที่เปลี่ยนแปลงคำสั่งเมื่อเงื่อนไขเปลี่ยน หรือการทำคำสั่งซ้ำๆ จำนวนมาก Control Flow จึงเกิดขึ้น 

Control Flow คือ การควบคุมการทำงานของคำสั่งต่างๆ อีกทีให้มีการทำงานร่วมกันอย่างเป็นระบบ ตามขั้นตอนที่ต้องการ

## Flow Chart แสดงภาพการทำงานของโปรแกรม

การอธิบายเป็นภาพโดยช่วยให้เราเข้าใจการทำงานที่ซับซ้อน เช่น การทำงานของโปรแกรมที่มีการแยกเงื่อนไขของการกระทำ บางครั้งทำแบบหนึ่ง บางครั้งทำอีกแบบ แล้วแต่เงื่อนไขจะเปลี่ยน ได้มากยิ่งกว่าการเขียนอธิบายอย่างเดียว ทำให้เราสามารถเลือกคำสั่ง เครื่องมือที่เหมาะสมสำหรับการเขียนคำสั่ง และในขณะเดียวกันช่วยเราติดตามผลของการสร้างโปรแกรมของเราอีกด้วย

## if statement แบ่งกรณีสั่งคำสั่ง

ในการสร้างเงื่อนไข ที่สั่งคำสั่งไม่เหมือนกันเมื่อเงื่อไขเปลี่ยนไป เราจะใช้ if statement เพื่อทำงานกับเงื่อนไข และจัดการสั่งคำสั่งตามที่เราต้องการตามสถานการณ์ที่กำหนด โดย if statement จะทำงานกับ boolean ที่เป็นเหมือน switch ที่คอยเปลี่ยนการทำงานตามเงื่อนไข

```python
number = # สร้างตัวเลขที่คุณต้องการ
```

```python
if number > 0:
    print(f'Your number, {number}, is a positive number.')
elif number < 0:
    print(f'Your number, {number}, is a negative number.')
else:
    print('Your number is zero!')
```

## While ทำซ้ำในขณะที่เป็นจริง

While เป็นการสร้างเงื่อนไขเพื่อการทำซ้ำโดยเฉพาะ ซึ่ง While จะทำงานกับ boolean และจะเริ่มทำงาน และทำซ้ำไปเรื่อยๆ เมื่อเงื่อนไขเป็น True และหยุดเมื่อเป็น False หรือใช้คำสั่งพิเศษ ดังนั้นหลักการสร้าง while จริงๆ แล้วประกอบด้วย 2 ส่วน คือ 1) ต้องออกแบบให้เงื่อนไขส่งค่า True เมื่อต้องการให้ while loop ทำงาน 2) ออกแบบให้เงื่อนไขส่งค่า False เมื่อไม่ต้องการให้ทำงาน ซึ่งทั้ง 2 ส่วนมีความสำคัญเท่ากันในการสั่งให้โปรแกรมทำงาน และหยุด กลายเป็น while loop ที่สมบูรณ์

```python
max_count = # ใส่จำนวนที่ต้องการ
```

```python
count = 0
while count < max_count:
    print(f'Counting... {count}')
    count += 1
else:
    print('Counting Completed!')
```
